#### 3.2.3: Checking for Outliers

```python
# Checking for Outliers in a numerical column ('Energy Consumption (kWh)
col = 'Energy Consumption (kWh)'

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

print(f"Number of outliers in '{col}': {outliers.shape[0]}")
print(outliers.head(3))
```

**Preliminary Insights:**<br>
A preliminary exploration of the dataset shows no missing or duplicate values. However, it reveals a significant presence of outliers (7997), approximately 8% of the 100,000 records. These outliers may stem from a variety of causes, such as:

**Erroneous temperature readings**, for example, extremely high temperatures during winter or unusually low temperatures in summer.<br>
**Abnormal energy consumption patterns**, which could could be from factors like illegal usage, faulty equipment, or sensor malfunctions.
Why Outliers Matter and also Pose Challenges in Energy Data:
Outliers in energy consumption data can be both important and problematic. They can:

- Skew statistical metrics like the mean and standard deviation, leading to inaccurate insights
- Obscure real anomalies or operational issues
- Reduce forecasting accuracy in smart grids
- Misrepresent consumption patterns due to sensor errors or rare events

Effectively handling outliers in energy data is a complex task, as removing them outright may discard valuable information, while retaining them can introduce noise and reduce analytical accuracy and model performance. 

**This section will take a deeper look into the temperature and energy consumption outliers.**

```python
# Convert 'Date' column to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Drop rows with invalid dates
df = df.dropna(subset=["Date"])

# Add a 'Season' column
def get_season(month):
    if month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    elif month in [9, 10, 11]:
        return "Autumn"
    else:
        return "Winter"

df["Season"] = df["Date"].dt.month.apply(get_season)

# Function to find outliers using 95th percentile threshold
def find_upper_outliers_95(data, column):
    threshold_95 = data[column].quantile(0.95)
    return data[data[column] > threshold_95]

# Find and store outliers by season
dataset_outliers_by_season = {}

for season in ["Spring", "Summer", "Autumn", "Winter"]:
    season_df = df[df["Season"] == season]
    outliers = find_upper_outliers_95(season_df, "Outdoor Temperature (°C)")
    dataset_outliers_by_season[season] = outliers
    print(f"{season}: {len(outliers)} outliers found")

# Example: View Spring outliers
print("\nDataset Temperature Outliers (Winter - above 95th percentile):")
print(dataset_outliers_by_season["Winter"].head(3))
```

#### 3.2.4: Adjust Temperature values against Outliers

Since the dataset has no defined location, using geolocation and weather APIs to obtain temperature data to replace temperature outliers is not an obtion. To obtain adjustment values of the "Outdoor Temperature (°C)" outliers, the seasonal mean was used. A new column was created in the dataset with outliers values replaced under the new column called "Outdoor Temperature 2 (°C)" which will be used for this project analysis.

- This approach uses the 95th percentile to define outliers and replaces them with the mean outdoor temperature of the respective season.
    -  The following codes were used to exercuted the task.

```python
# Convert 'Date' column to datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Date"])

# Generate Hour if not present
if "Hour" not in df.columns and "Time" in df.columns:
    df["Hour"] = pd.to_datetime(df["Time"], format="%H:%M", errors="coerce").dt.hour

# Create Time Segment if not already there
if "Time Segment" not in df.columns:
    df["Time Segment"] = pd.cut(
        df["Hour"],
        bins=[-1, 5, 11, 17, 23],
        labels=["Night", "Morning", "Afternoon", "Evening"] )

# Define season function
def get_season(month):
    if month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    elif month in [9, 10, 11]:
        return "Autumn"
    else:
        return "Winter"

# Apply season
df["Season"] = df["Date"].dt.month.apply(get_season)

# Make a copy of the temperature column to adjust
df["Outdoor Temperature 2 (°C)"] = df["Outdoor Temperature (°C)"]

# Iterate over seasons
for season in ["Spring", "Summer", "Autumn", "Winter"]:
    # Subset season data
    season_mask = df["Season"] == season
    season_data = df[season_mask]

    # 95th percentile threshold for temperature (season-wide)
    threshold = season_data["Outdoor Temperature (°C)"].quantile(0.95)

    # Identify season outliers
    outlier_mask = season_mask & (df["Outdoor Temperature (°C)"] > threshold)

    # Now replace per time segment
    for segment in ["Morning", "Afternoon", "Evening", "Night"]:
        # Define time mask
        time_mask = df["Time Segment"] == segment

        # Filter for this season & time segment & non-outliers to calculate mean
        non_outlier_mean = df[
            season_mask & time_mask & (df["Outdoor Temperature (°C)"] <= threshold)
        ]["Outdoor Temperature (°C)"].mean()

        # Find outliers for this group
        replace_mask = outlier_mask & time_mask

        # Replace with time-segment specific seasonal mean
        df.loc[replace_mask, "Outdoor Temperature 2 (°C)"] = non_outlier_mean

print(" Outdoor Temperature outliers adjusted per season and time segment.")
```

<font color=blue>**Question 3:** ➤</font> Show the distribution of energy consumption at household level using scatterplot.

```python
# Ensure correct data types
df["Home ID"] = df["Home ID"].astype(int)
df["Household Size"] = df["Household Size"].astype(int)

# Aggregate total energy consumption per Home ID
total_consumption = (
    df.groupby(["Home ID", "Household Size"])["Energy Consumption (kWh)"]
    .sum()
    .reset_index()
    .rename(columns={"Energy Consumption (kWh)": "Total Consumption (kWh)"}))

# Create color mapping for household size
import seaborn as sns
unique_sizes = sorted(total_consumption["Household Size"].unique())
palette = sns.color_palette("tab10", n_colors=len(unique_sizes))
color_map = {size: palette[i] for i, size in enumerate(unique_sizes)}

# Assign colors
total_consumption["Color"] = total_consumption["Household Size"].map(color_map)

# Plot
plt.figure(figsize=(10, 6))
for size in unique_sizes:
    subset = total_consumption[total_consumption["Household Size"] == size]
    plt.scatter(
        subset["Home ID"],
        subset["Total Consumption (kWh)"],
        label=f"{size} person(s)",
        color=color_map[size],
        edgecolor="black",
        alpha=0.8,
        s=100)

# Customize plot
plt.title("Total Household Energy Consumption by Household Size", fontsize=15)
plt.xlabel("Home ID", fontsize=12)
plt.ylabel("Total Consumption (kWh)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Household Size", bbox_to_anchor=(1.01, 1), loc="upper left", borderaxespad=0.)
plt.tight_layout()
plt.show()
```

<font color=blue>**Question 10:** ➤</font> How does temperature impact energy consumption or How does energy consumption change with outdoor temperature (°C)?

```python
# Clean data: drop rows with missing values in relevant columns
df_clean = df.dropna(subset=["Outdoor Temperature 2 (°C)", "Energy Consumption (kWh)"])

# Define temperature categories
def temp_category(temp):
    if temp >= 30:
        return "≥30°C (Hot)"
    elif temp < 0:
        return "<0°C (Cold)"
    else:
        return "0-29°C (Moderate)"

df_clean["Temp Category"] = df_clean["Outdoor Temperature 2 (°C)"].apply(temp_category)

# Aggregate average energy consumption by temperature (rounded)
df_clean["Temp Rounded"] = df_clean["Outdoor Temperature 2 (°C)"].round()

temp_energy = (
    df_clean.groupby(["Temp Rounded", "Temp Category"], observed=True)["Energy Consumption (kWh)"]
    .mean()
    .reset_index())

# Plot lineplot showing energy consumption by temperature with categories
plt.figure(figsize=(12, 6))
sns.lineplot(data=temp_energy, x="Temp Rounded", y="Energy Consumption (kWh)", hue="Temp Category", palette="coolwarm")

plt.title("Energy Consumption vs Outdoor Temperature with Extreme Temperature Categories")
plt.xlabel("Outdoor Temperature (°C)")
plt.ylabel("Average Energy Consumption (kWh)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Temperature Category")
plt.tight_layout()
plt.show()

# Optional: print average consumption for each category
avg_consumption_by_cat = df_clean.groupby("Temp Category")["Energy Consumption (kWh)"].mean()
print("Average Energy Consumption by Temperature Category:")
print(avg_consumption_by_cat, '\n')

# Simple correlation between temperature and energy use
correlation = df["Outdoor Temperature 2 (°C)"].corr(df["Energy Consumption (kWh)"])
print(f"Correlation between Outdoor Temperature and Energy Consumption: {correlation:.4f}")
```

**Interpretation:**
- The result shows a extremely maginal possitive correlation between Outdoor Temperature and Energy Consumption with a coefficient 0.0014, that is close to 0. Indicating no linear relationship between outdoor temperature and energy consumption.
- Therefore outdoor temperature is not a predictable effect of energy consumption in a linear manner.

*Possible Reasons:*
- Appliance usage patterns (behavioral) may dominate temperature effects (e.g., heating, cooling, lighting, cooking, etc.).
- Dataset might be collected from a single region where Housing insulation is predominant in the construction style or having HVAC systems that may reduce temperature impact.

```python
# Group by Home ID and calculate total and outlier counts
household_outliers = df.groupby("Home ID")["High Energy Flag"].agg(
    Total_Records="count",
    Outlier_Count="sum").reset_index()

# Convert columns to numeric (safeguard against strings)
household_outliers["Total_Records"] = pd.to_numeric(household_outliers["Total_Records"], errors="coerce")
household_outliers["Outlier_Count"] = pd.to_numeric(household_outliers["Outlier_Count"], errors="coerce")

# Calculate the outlier rate
household_outliers["Outlier Rate (%)"] = (household_outliers["Outlier_Count"] / household_outliers["Total_Records"]) * 100

# Sort by highest outlier rate
household_outliers = household_outliers.sort_values("Outlier Rate (%)", ascending=False)

# Display top households
print("=== Households with Highest Outlier Rates:")
print(household_outliers.head(10))

# Filter for households with more than 20% outliers and at least 10 outlier records
consistent_outliers = household_outliers[
    (household_outliers["Outlier Rate (%)"] > 5) &
    (household_outliers["Outlier_Count"] >= 15)]

print("\n=== Consistently Anomalous Households ===")
print(consistent_outliers.head(5).round(2))
```

<font color=blue>**Question 22:** ➤</font> Do any homes show unusually high or low consumption compared to others in the same size category?

```python
# Step 1: Calculate average energy consumption per Home ID
home_avg = df.groupby(['Home ID', 'Household Size'])['Energy Consumption (kWh)'].mean().reset_index()

# Step 2: Calculate mean energy consumption per household size
home_avg['Size Mean Consumption'] = home_avg.groupby('Household Size')['Energy Consumption (kWh)'].transform('mean')

# Step 3: Calculate % deviation from group mean
home_avg['Deviation (%)'] = ((home_avg['Energy Consumption (kWh)'] - home_avg['Size Mean Consumption']) / 
                             home_avg['Size Mean Consumption']) * 100

# Step 4: Flag outlier type
def label_outlier(row):
    if row['Deviation (%)'] > 25:
        return 'High'
    elif row['Deviation (%)'] < -25:
        return 'Low'
    else:
        return 'Normal'

home_avg['Outlier Type'] = home_avg.apply(label_outlier, axis=1)

# Step 5: Filter only unusual homes
unusual_homes = home_avg[home_avg['Outlier Type'] != 'Normal'].copy()

# Step 6: Format columns
unusual_homes['Energy Consumption (kWh)'] = unusual_homes['Energy Consumption (kWh)'].round(2)
unusual_homes['Size Mean Consumption'] = unusual_homes['Size Mean Consumption'].round(2)
unusual_homes['Deviation (%)'] = unusual_homes['Deviation (%)'].round(2)

# Step 7: Final table
unusual_table = unusual_homes[['Home ID', 'Household Size', 'Energy Consumption (kWh)',
                               'Size Mean Consumption', 'Deviation (%)', 'Outlier Type']] \
                  .sort_values(by='Deviation (%)', ascending=False)

# Step 8: Print first 5 and last 5 records
preview_table = pd.concat([unusual_table.head(5), unusual_table.tail(5)])
print("=== Homes with Unusual Energy Consumption by Household Size (First & Last 5) ===")
print(preview_table.to_string(index=False))

# Step 9: Show total number of unique Home IDs in the table
unique_outlier_count = unusual_table['Home ID'].nunique()
print(f"\nTotal unique Home IDs with unusual consumption: {unique_outlier_count}")
```

```python
# Ensure home-level averages by household size are available
home_avg = df.groupby(['Home ID', 'Household Size'])['Energy Consumption (kWh)'].mean().reset_index()

# Compute mean energy consumption per household size
group_means = home_avg.groupby('Household Size')['Energy Consumption (kWh)'].transform('mean')

# Calculate deviation (%) from the household size group mean
home_avg["Deviation (%)"] = ((home_avg["Energy Consumption (kWh)"] - group_means) / group_means) * 100

# Plot the distribution of deviations
plt.figure(figsize=(10, 6))
sns.histplot(home_avg["Deviation (%)"], kde=True, bins=30, color='steelblue')

# Add threshold lines
plt.axvline(50, color="red", linestyle="--", label="Upper Threshold: +50%")
plt.axvline(-50, color="red", linestyle="--", label="Lower Threshold: -50%")

# Titles and labels
plt.title("Distribution of Energy Usage Deviation (%)\nfrom Group Mean by Household Size")
plt.xlabel("Deviation from Group Mean (%)")
plt.ylabel("Number of Homes")
plt.legend()
plt.tight_layout()
plt.show()
```

<font color=blue>**Question 23:** ➤</font> How can energy consumption outliers help detect illegal or suspicious consumption?

Solution:
- Set a threshold limit of deviation of Household consumption from average: example Home ID where Energy Consumption (kWh) has over 75% deviation from their mean Energy Consumption (kWh) levels might be illegal consumption or faulty smart meter sensor / equipment.
- Utility companies may charge household with consumption within the normal range lesser as opposed to consumption above the normal range.

```python
# Drop existing Mean_Consumption column if it exists
if "Mean_Consumption" in df.columns: df = df.drop(columns=["Mean_Consumption"])

# Step 1: Calculate mean energy consumption per Home ID
home_means = df.groupby("Home ID")["Energy Consumption (kWh)"].mean().rename("Mean_Consumption")

# Step 2: Merge mean back into the original dataframe
df = df.merge(home_means, on="Home ID")

# Step 3: Calculate percentage deviation from mean
df["% Deviation"] = ((df["Energy Consumption (kWh)"] - df["Mean_Consumption"]).abs() / df["Mean_Consumption"]) * 100

# Step 4: Flag suspicious records where deviation is over 75%
df["Suspicious Flag"] = np.where(df["% Deviation"] > 75, 1, 0)

# Step 5: Aggregate suspicious results by Home ID
suspicious_summary = (df[df["Suspicious Flag"] == 1].groupby("Home ID").agg(
        Total_Records=("Home ID", "count"),
        Mean_Consumption=("Mean_Consumption", "mean"),
        Avg_Deviation_Percent=("% Deviation", "mean")).reset_index())

# Add Suspicious Flag column
suspicious_summary["Suspicious Flag"] = 1

# Round for better display
suspicious_summary["Mean_Consumption"] = suspicious_summary["Mean_Consumption"].round(2)
suspicious_summary["Avg_Deviation_Percent"] = suspicious_summary["Avg_Deviation_Percent"].round(2)

# Sort by highest average deviation
suspicious_summary = suspicious_summary.sort_values(by="Avg_Deviation_Percent", ascending=False)

print(f"=== TOTAL SUSPECT RECORDS (>75% deviation): {len(suspicious_summary)} ===")
print(suspicious_summary.head(10).to_string(index=False))

# Visualize deviation distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["% Deviation"], kde=True, bins=30, color='skyblue')
plt.axvline(75, color="red", linestyle="--", label="Suspicious Threshold: 75%")
plt.title("Distribution of Energy Usage Deviation (%) from Expected Mean")
plt.xlabel("Deviation (%)")
plt.ylabel("Number of Records")
plt.legend()
plt.tight_layout()
plt.show()
```

**Interpretation and other insights:**

The graph is a bimodal, right-skewed distribution of the same dataset, meaning:
- There are two distinct peaks in the distribution, indicating two patterns in the data.
- The first peak representing typical users and the second peak representing extreamely high consumers.
- Deviations that falles between -50% and +50% from mean consumption values falls under the normal range, above 50% to 74% falls under high energy users, and above 75% deviation from mean consumption depending on the threshold limit set by the utility company are extremely high energy users deemed suspicious.
- Utility companies may decide to charge normal range users less than high energy users per kWh of energy consumption as their consumption levels put lesser stress on the energy infracstructures particularly during peak load.
  
The chart shows that, the mean consumption appears normal at first glance (1.5 kWh), but a second peak emerges, accompanied by a significantly inflated percentage deviation. This second peak extends into a long rightward tail, indicating inconsistent usage patterns within individual records.

- Deviations above the suspicious threshold are usually driven by or indicates the following:
    - Malfunctioning or misreported sensors/meters due technical issues or weather extremes
    - Unusual usage behavior (e.g., appliances left running)
    - Potential energy theft or system errors
      
This analysis helps flag homes that need further investigation, possibly through audits, maintenance checks, or user interviews.

### 4.7: Advanced Insights and Predictive Modeling:
   Use statistical and machine learning approaches to identify key drivers of high energy usage and segment behavior.

<font color=blue>**Question 25:** ➤</font> Which features (appliance, temperature, time, etc.) contribute most to predicting high energy consumption?

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# Filter necessary columns and drop rows with missing values
features = [
    "Appliance Type", "Outdoor Temperature (°C)", "Time Segment",
    "Season", "Household Size"]
df_model = df[features + ["High Energy Flag"]].dropna()

# Encode categorical features
le_appliance = LabelEncoder()
le_time_segment = LabelEncoder()
le_season = LabelEncoder()

df_model["Appliance Type"] = le_appliance.fit_transform(df_model["Appliance Type"])
df_model["Time Segment"] = le_time_segment.fit_transform(df_model["Time Segment"])
df_model["Season"] = le_season.fit_transform(df_model["Season"])

# Split into features and target
X = df_model.drop("High Energy Flag", axis=1)
y = df_model["High Energy Flag"]

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Get feature importances
importances = model.feature_importances_
feature_names = X.columns
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance Score": importances
}).sort_values(by="Importance Score", ascending=True).reset_index(drop=True)

# Show result
print("=== Feature Importance for Predicting High Energy Consumption ===")
print(importance_df.round(4), '\n')

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    x="Importance Score",
    y="Feature", hue="Importance Score", 
    data=importance_df.sort_values("Importance Score", ascending=False),
    palette="viridis")

plt.title("Feature Importance for Predicting High Energy Consumption")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.legend().remove()
plt.show()
```

**Interpretation:**
- The result shows features/factors that contribute most to predicting high energy consumption.
- High Energy consumption predictors in order of magnitude:
    - *(1) Outdoor Temperature, (2) Energy Consumption, (3) Appliance Type, (4) Household Size, (5) Time Segment, (6) Season*.

<font color=blue>**Question 26:** ➤</font> Predict next month consumption of Home IDs 1 to 5 based on past consumption records.

**Strategy:**
Here a basic linear trend model will be used (e.g., linear regression) to predict future values per Home ID using their historical monthly consumption.

**The Step-by-Step Prediction Plan:**
1. Filter data for Home IDs 1–5
2. Aggregate monthly energy usage per Home ID
3. Convert the time period to numeric values (for modeling)
4. Fit a linear regression for each Home ID
5. Predict next month’s consumption

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Ensure datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Filter Home IDs 1 to 5
df_filtered = df[df["Home ID"].isin([1, 2, 3, 4, 5])].copy()

# Create YearMonth column
df_filtered["YearMonth"] = df_filtered["Date"].dt.to_period("M")
df_filtered["YearMonth_dt"] = df_filtered["YearMonth"].dt.to_timestamp()
df_filtered["Month_Num"] = df_filtered["YearMonth_dt"].dt.to_period("M").apply(lambda x: x.ordinal)

# Monthly energy consumption per home
monthly_data = (
    df_filtered.groupby(["Home ID", "YearMonth", "Month_Num"])["Energy Consumption (kWh)"]
    .sum()
    .reset_index())

# Determine latest available month in the dataset
latest_month_num = monthly_data["Month_Num"].max()
next_month_num = latest_month_num + 1
next_month_label = pd.Period(year=pd.Timestamp.now().year, month=pd.Timestamp.now().month, freq="M") + 1

# Build the forecast table
forecast_records = []

for home_id in range(1, 6):
    home_df = monthly_data[monthly_data["Home ID"] == home_id]
    
    if len(home_df) < 2:
        continue  # Not enough data to fit a trend
    
    # Prepare data for model
    X = home_df["Month_Num"].values.reshape(-1, 1)
    y = home_df["Energy Consumption (kWh)"].values
    model = LinearRegression()
    model.fit(X, y)
    
    # Forecast next month
    predicted_kwh = round(model.predict(np.array([[next_month_num]]))[0], 2)
    
    # Get past month's consumption (latest available)
    past_kwh = round(home_df[home_df["Month_Num"] == latest_month_num]["Energy Consumption (kWh)"].values[0], 2)
    
    # Append result
    forecast_records.append({
        "Home ID": home_id,
        "Predicted Month": next_month_label.strftime("%Y-%m"),
        "Past Consumption (kWh)": past_kwh,
        "Predicted Consumption (kWh)": predicted_kwh})

# Create DataFrame for display
forecast_df = pd.DataFrame(forecast_records)

# Display neatly
print("=== Forecasted Energy Consumption for Next Month (Based on Trend) ===")
print(forecast_df.to_string(index=False))
```

### 6.1: Limitations of the Study

**Limitations of the Study**
1. **Granularity and Frequency of Data:** Although the dataset seems to record data on a daily or hourly basis, it might not offer real-time measurements, which makes it more difficult to identify transient consumption abnormalities or spikes.

2. **Missing location:** Outliers were found in outdoors temperature reading but with missing location, geolocation and weather APIs could not be used to substituted the actual temperature data. Mean values were used for each corresponding time segment which migh have had an influence on the analysis.

3. **Lack of Contextual Features:** There are missing variables such as building type, energy tariffs, income level, and behavioral aspects (e.g., occupancy, working hours). Although they are not taken into consideration, these factors can have a substantial impact on energy use.

4. **Limited Appliance-Level Monitoring:** While energy usage by appliance is provided, disaggregated or smart appliance usage data (e.g., IoT device logs) is likely limited or estimated, reducing the accuracy of appliance specific insights.

5. **Potential Inaccuracies in Data Entry:** Problems like inconsistent formats, duplication, or missing values (particularly in datetime fields) point to a lack of automated validation throughout the data gathering phase.

6. **Bias in Household Representation:** The results may not apply to all households if the majority of the residences in the dataset are concentrated in a specific area, climate zone, or socioeconomic class.

7. **Static Data (Not Real-Time):** The dataset covers a set time period (January 2023 to January 2024) and does not include real-time dynamics or changing consumption patterns (such as the deployment of solar panels or smart thermostats).
